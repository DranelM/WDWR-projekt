```python
import pandas as pd
import numpy as np

scenarios = pd.read_csv('scenarios.csv')
scenarios.head()
```

```python
scenarios['R7'] = pd.Series(0.5, index=scenarios.index)
```

```python
scenarios['R8'] = pd.Series(0.3, index=scenarios.index)
```

```python
scenarios.head()
```

```python
min_koszt_d = np.array([120, 0,10,0,0,90,0,100])
min_ryzyko_d = np.array([0,130,0,90,0,0,0,0])
posredni_d = np.array([30,100,0,90,0,0,0,0])
```

```python
# min_koszt = min_koszt_d * scenarios.val
min_koszt = np.sum(scenarios.values * min_koszt_d, axis=1)
min_ryzyko = np.sum(scenarios.values * min_ryzyko_d, axis=1)
posredni = np.sum(scenarios.values * posredni_d, axis=1)
```

```python
min_koszt = np.around(min_koszt, 4)
```

```python
min_koszt
```

```python
df = pd.DataFrame({"min_koszt": min_koszt, "min_ryzyko": min_ryzyko.T, "posredni":posredni})
```

```python
df.head()
```

```python
df.to_csv("dystrybuanty.csv", index=None)
```
